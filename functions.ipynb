```python
def say_hello():
    print("Hello")
    
say_hello()
print()

### Parameters ###

def say_hello_to(name):
    print(f"Hello, {name}!")

say_hello_to("Alice")
print()

### Default Parameters and Keyword Arguments ###

def introduce_yourself(name, age):
    print(f"My name is {name} and I am {age} years old.")

introduce_yourself("Bob", 30) # Positional arguments
introduce_yourself(age=25, name="Charlie") # Keyword arguments

```
