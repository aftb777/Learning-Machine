0. Import matplotlib and get it ready for plotting in Jupyter


```python
%matplotlib inline
import matplotlib.pyplot as plt
```

1. Prepare data

```python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
```

2. Setup plot

```python
fig, ax = plt.subplots(figsize=(10, 10))  # (width, height)
```

3. Plot data

```python
ax.plot(x, y)
```

4. Customize plot

```python
ax.set(
    title="Simple Plot",
    xlabel="x-axis",
    ylabel="y-axis"
)
```
