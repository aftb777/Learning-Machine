```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
car_sales = pd.read_csv("/Users/aftaabmulla/DS-ML-AI/PandasCodes/car-sales.csv")
car_sales
```

Cummulative Sum of Pandas Data

```python
time_stamp = pd.Series(np.random.randn(1000), 
                       index=pd.date_range('1/1/2026', periods=1000))

time_stamp = time_stamp.cumsum()

time_stamp.plot(
    title="Cumulative Sum",
    xlabel="Numbers",
    ylabel="Dates"
);
```

Plotting 2

```python
# Basic Plot
car_sales.plot()
plt.show()
```

```python
car_sales["Price"] = (
    car_sales["Price"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
)

car_sales["Price"] = pd.to_numeric(car_sales["Price"])
```

```python
#line plot
car_sales["Price"].plot()

# car_sales.plot(x="Make", y="Price")
```

```python
# Bar Plot
car_sales.plot(kind="bar")

# car_sales.plot(x="Make", y="Price", kind="bar")
```

```python
# Horizontal Bar
car_sales.plot(kind="barh")
```

```python
# Histogram
car_sales["Price"].plot(kind="hist")

# car_sales["Price"].hist()
```

```python
# Scatter Plot
car_sales.plot(
    x="Odometer (KM)",
    y="Price",
    kind="scatter"
)
```
