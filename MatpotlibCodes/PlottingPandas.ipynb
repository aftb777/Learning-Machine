Cummulative Sum of Pandas Data

```python
time_stamp = pd.Series(np.random.randn(1000), 
                       index=pd.date_range('1/1/2026', periods=1000))

time_stamp = time_stamp.cumsum()

time_stamp.plot(
    title="Cumulative Sum",
    xlabel="Numbers",
    ylabel="Dates"
);
```
