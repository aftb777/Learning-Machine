# Plotting from Pandas Data Frame

Plotting 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
car_sales = pd.read_csv("/Users/aftaabmulla/DS-ML-AI/PandasCodes/car-sales.csv")
car_sales
```

Cummulative Sum of Pandas Data

```python
time_stamp = pd.Series(np.random.randn(1000), 
                       index=pd.date_range('1/1/2026', periods=1000))

time_stamp = time_stamp.cumsum()

time_stamp.plot(
    title="Cumulative Sum",
    xlabel="Numbers",
    ylabel="Dates"
);
```

Plotting 2

```python
# Basic Plot
car_sales.plot()
plt.show()
```

```python
car_sales["Price"] = (
    car_sales["Price"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
)

car_sales["Price"] = pd.to_numeric(car_sales["Price"])
```

```python
#line plot
car_sales["Price"].plot()

# car_sales.plot(x="Make", y="Price")
```

```python
# Bar Plot
car_sales.plot(kind="bar")

# car_sales.plot(x="Make", y="Price", kind="bar")
```

```python
# Horizontal Bar
car_sales.plot(kind="barh")
```

```python
# Histogram
car_sales["Price"].plot(kind="hist");

# car_sales["Price"].hist()
```

```python
# Scatter Plot
car_sales.plot(
    x="Odometer (KM)",
    y="Price",
    kind="scatter"
);
```

Plotting 3

```python
# Box Plot
car_sales["Price"].plot(kind="box");
```

```python
# Area plot
car_sales["Price"].plot(kind="area");
```

```python
# Pie chart
car_sales.Price.plot(kind="pie");
```

```python
# Density Plot
car_sales.Price.plot(kind="density");
```

```python
# Hexbian Plot
car_sales.plot(
    x="Doors",
    y="Price",
    kind="hexbin"
);
```

```python
car_sales["Total_sales"] = car_sales["Price"].cumsum()

car_sales.plot(
    x="Make",
    y="Total_sales"
);
```

Plotting 4

```python
# Figure Size
car_sales.plot(figsize=(10, 5));
```

```python
# Title
car_sales.plot()
plt.title("Car Sales");
```

```python
# Lables
car_sales.plot(
    x="Make",
    y="Price"
);

plt.xlabel("Company Name")
plt.ylabel("Price of Car");
```

```python
# Legend - Inbuild
car_sales.plot()

# optional 
plt.legend();
```

```python
# Grid
car_sales.plot()
plt.grid(True);
```

```python
# Style
car_sales.plot()
plt.style.use("classic")

plt.style.available
```
