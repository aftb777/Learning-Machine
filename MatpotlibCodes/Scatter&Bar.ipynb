```python
import numpy as np

arr = np.linspace(0, 25, num=100)

print(arr)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(
    arr, arr**2
)
```

# Use Same data and create Scatter plot

```python
fig , ax = plt.subplots()
ax.scatter(
    arr, np.sin(arr) ## Sin wave
)
```

# Scatter plot with dictionaries 

```python
butter_prices = {"Almond" : 15,
                 "Peanut" : 10,
                 "Cashew" : 25}
```

```python
fig , ax = plt.subplots()
ax.plot(
    butter_prices.keys(),
    butter_prices.values()
)
```

```python
fig , ax = plt.subplots()
ax.scatter(
    butter_prices.keys(),
    butter_prices.values()
)
```

```python
fig , ax = plt.subplots()
ax.barh(
    list(butter_prices.keys()),
    list(butter_prices.values())
)
```

```python
fig , ax = plt.subplots()
ax.bar(
    butter_prices.keys(),
    butter_prices.values()
);
```
