```python
import numpy as np
import matplotlib.pyplot as plt
```

# Histogram

```python
arr = np.random.randn(1000)
```

```python
fig, ax = plt.subplots()
ax.hist(
    arr
);
```

```python
butter_prices = {"Almond" : 15,
                 "Peanut" : 10,
                 "Cashew" : 25}
```

# 2 Options for Subplot

```python
# Option-1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

for ax in [ax1, ax2, ax3, ax4]:
    ax.grid(True)

ax1.plot(arr, arr*2);
ax2.scatter(np.random.random(100), np.random.random(100));
ax3.bar(butter_prices.keys(), butter_prices.values());
ax4.hist(butter_prices.values());
```

```python
# Options 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

ax[0, 0].plot(arr, arr*2);
ax[0, 1].scatter(np.random.random(100), np.random.random(100));
ax[1, 0].bar(butter_prices.keys(), butter_prices.values());
ax[1, 1].hist(butter_prices.values());
```
