```python
dictionary = {'a' : 1, 'b' : 2, 'c' : 3}
print(dictionary['a'])


new_dict = {'a' : [1, 2, 3],
            'b' : "hello",
            'c' : True
            }

## Key is immutable, value can be mutable or immutable
print(new_dict['a'])


### Dictionary Methods ###
print(new_dict.keys())
print(new_dict.values())
print(new_dict.items())

print(new_dict.get('b'))
print(new_dict.get('d', 'Not Found'))  # Default value if key is not
print(new_dict['d'])  # This will raise a KeyError

new_dict2 = new_dict.copy()  # This will create a shallow copy of the dictionary
new_dict.clear()  # This will clear all items from the dictionary
print(new_dict)  # This will print an empty dictionary
print(new_dict2)  # This will print the copied dictionary with original items
```
