```python
# Import missing data
car_sales_missing = pd.read_csv("PandasCodes/car-sales-missing-data.csv")

print(car_sales_missing)
```

```python
# Fill missing Odometer values
car_sales_missing["Odometer"].fillna(
    car_sales_missing["Odometer"].mean(),
    inplace=True
)

print(car_sales_missing)
```

```python
# Remove remaining missing values
car_sales_missing.dropna(inplace=True)

print(car_sales_missing)
```

```python
# Seats column
car_sales_missing["Seats"] = 5
```

```python
# Engine Size column
car_sales_missing["Engine Size"] = np.random.uniform(
    1.3,
    4.5,
    len(car_sales_missing)
).round(1)
```

```python
# Price per KM
car_sales_missing["Price per KM"] = (
    car_sales_missing["Price"] /
    car_sales_missing["Odometer"]
)

print(car_sales_missing)
```

```python
# Remove last column
car_sales_missing.drop(
    "Price per KM",
    axis=1,
    inplace=True
)
```

```python
# Shuffle DataFrame
shuffled = car_sales_missing.sample(frac=1)

print(shuffled)
```

```python
# Reset index
shuffled = shuffled.reset_index(drop=True)

print(shuffled)
```

```python
# Convert KM to Miles
shuffled["Odometer (Miles)"] = shuffled["Odometer"] * 0.621371
```

```python
# Remove KM column
shuffled.drop("Odometer", axis=1, inplace=True)

print(shuffled)
```
