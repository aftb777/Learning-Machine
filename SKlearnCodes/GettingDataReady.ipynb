# Feature Encoding using OneHotEncoder

## What is Feature Encoding?

Feature Encoding is a preprocessing technique used to convert **categorical (text) data** into **numerical values**, because Machine Learning algorithms can only work with numbers.

---

## Step 1: Import Required Libraries

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

### OneHotEncoder

`OneHotEncoder` converts categorical values into binary (0 and 1) columns.

Example:

| Colour |
|---------|
| Red |
| Blue |
| Green |

After One Hot Encoding:

| Colour_Blue | Colour_Green | Colour_Red |
|--------------|--------------|------------|
| 0 | 0 | 1 |
| 1 | 0 | 0 |
| 0 | 1 | 0 |

---

## Step 2: Select Categorical Features

```python
categorical_features = ["Make", "Colour", "Doors"]
```

These are the columns that need to be encoded.

Only these columns will be transformed into numerical values.

---

## Step 3: Create OneHotEncoder Object

```python
one_hot = OneHotEncoder()
```

At this stage, no encoding is performed.

The encoder object is simply created and is ready to learn the categories.

It is similar to:

```python
model = RandomForestClassifier()
```

The model is created but has not been trained yet.

---

## Step 4: Create a ColumnTransformer

```python
transformer = ColumnTransformer(
    transformers=[
        ("one_hot", one_hot, categorical_features)
    ],
    remainder="passthrough"
)
```

### Why ColumnTransformer?

A dataset may contain both categorical and numerical columns.

Example:

| Make | Colour | Doors | Odometer |
|------|---------|-------|----------|
| Toyota | Red | 4 | 150000 |

Only **Make**, **Colour**, and **Doors** should be encoded.

The **Odometer** column is already numerical, so it should remain unchanged.

`ColumnTransformer` allows different transformations to be applied to different columns.

---

### Understanding the tuple

```python
("one_hot", one_hot, categorical_features)
```

- `"one_hot"` → Name of the transformer (can be anything)
- `one_hot` → The OneHotEncoder object
- `categorical_features` → List of columns to encode

---

## What does `remainder="passthrough"` mean?

```python
remainder="passthrough"
```

This tells the transformer:

> Keep all remaining columns unchanged.

Example:

Encoded Columns

- Make
- Colour
- Doors

Unchanged Columns

- Odometer
- Price

If you use

```python
remainder="drop"
```

all remaining columns will be removed from the output.

---

## Step 5: Apply Encoding

```python
transformed_X = transformer.fit_transform(X)
```

`fit_transform()` performs two operations.

### fit()

- Learns the unique categories from the dataset.

Example:

```
Red
Blue
Green
```

The encoder remembers these categories.

---

### transform()

- Converts the learned categories into binary (0/1) columns.

Example:

| Colour_Blue | Colour_Green | Colour_Red |
|--------------|--------------|------------|
| 0 | 0 | 1 |
| 1 | 0 | 0 |
| 0 | 1 | 0 |

The output is now completely numerical.

---

## Step 6: Display the Encoded Dataset

```python
transformed_X
```

This displays the transformed dataset after One Hot Encoding.

---

---

# Interview Notes

## OneHotEncoder

- Converts categorical features into binary (0/1) columns.
- Prevents Machine Learning models from assuming any order between categories.
- Commonly used for **nominal categorical data**.

---

## ColumnTransformer

- Applies different preprocessing techniques to different columns.
- Allows encoding of selected columns while leaving the remaining columns unchanged.
- `remainder="passthrough"` keeps the remaining columns.
- `remainder="drop"` removes the remaining columns.

---

## fit()

- Learns the unique categories from the training data.
- No transformation is performed during this step.

---

## transform()

- Converts categorical values into encoded numerical values using the learned categories.

---

## fit_transform()

Equivalent to:

```python
fit()
transform()
```

Both operations are performed in a single step.

---

# Key Takeaways

- Machine Learning models cannot understand text directly.
- OneHotEncoder converts categorical data into binary numerical columns.
- ColumnTransformer controls **which columns** should be transformed.
- `fit()` learns the categories.
- `transform()` applies the encoding.
- `fit_transform()` performs both operations together.

# Handling Missing values

1. Fill them with some value also called as Imputation
2. Remove samples with missing data

````python
car_missing = pd.read_csv("/Users/aftaabmulla/DS-ML-AI/DataSets/car-sales-extended-missing-data.csv")
````

````python
# check for missing numbers from each column
car_missing.isna().sum()
````

1. Option 1 :- fill with Pandas

````python
# Fill make column
car_missing["Make"] = car_missing["Make"].fillna("missing")

# Fill colour column
car_missing["Colour"] = car_missing["Colour"].fillna("missing")

# Fill odometer column
car_missing["Odometer (KM)"] = car_missing["Odometer (KM)"].fillna(
    car_missing["Odometer (KM)"].mean()
)

# Fill doors column
car_missing["Doors"] = car_missing["Doors"].fillna(4);
````
