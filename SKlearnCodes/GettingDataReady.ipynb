```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
heart = pd.read_csv("heart-disease.csv")
```

Step 1

```python
X = heart.drop("target", axis=1)
```

```python
Y = heart["target"]
```

```python
from sklearn.model_selection import train_test_split as tts
X_train, Y_train, X_test, Y_test = tts(
    X, Y,
    random_state=42,
    test_size=0.2
)
```
