```python
## Truthy and Falsy Values

# In Python, certain values are considered "truthy" or "falsy" when evaluated in a boolean context.
# Truthy values are those that evaluate to True, while falsy values evaluate to False.
# Falsy values include:
# - None
# - False
# - 0 (zero of any numeric type)
# - 0.0 (zero float)
# - 0j (zero complex)
# - Empty sequences and collections (e.g., '', [], {}, set())
# - Objects of user-defined classes that implement a __bool__() or __len__()
#   method that returns False or 0.
# All other values are considered truthy.

# Example usage:
# Falsy values
print(bool(None))  # Output: False
print(bool(False))  # Output: False
print(bool(0))  # Output: False
print(bool(0.0))  # Output: False

# Truthy values
print(bool(1))  # Output: True
print(bool(-1))  # Output: True
print(bool("Hello"))  # Output: True
print(bool([1, 2, 3]))  # Output: True
print(bool({'key': 'value'}))  # Output: True
```
