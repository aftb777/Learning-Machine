```python
lists1 = [1, 2, 4, 'a', 'b', 'c', True, False, 1.2]

print(lists1)

print(lists1[0])  # 1
print(lists1[3])  # 'a'
print(lists1[-1])  # 1.2
print(lists1[-3])  # True

### Slicing
print(lists1[0:4:1]) # [start:stop:stepover]

### List Methods
len(lists1)  # 9

lists1.append('new')  # Adds 'new' to the end of the list
print(lists1)

lists1.sort()  # Sorts the list in place
print(lists1)  # This will raise an error because the list contains different data types that cannot be compared for sorting
```
