```python
# Fundamental Data Types 

# Int 
x = 10
print(x)  # Output: 10
print(type(x))  # Output: <class 'int'>

print(2 + 3)  # Output: 5
print(2 * 3)  # Output: 6
print(10 / 2)  # Output: 5.0
print(3 - 2) # Output: 1

# Float
y = 3.14
print(y)  # Output: 3.14
print(type(2 / 4)) # Output: <class 'float'>

# Power operator
print(2 ** 3)  # Output: 8
print(5 // 4)  # Output: 1 (floor division- round down to nearest whole number)

# Modulo 
print(10 % 3)  # Output: 1 (remainder of 10 divided by 3)

print(bin(5))
```
