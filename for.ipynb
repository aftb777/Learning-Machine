```python
for i in "Aftaab":
    print(i)
    
print()

for i in [1, 2, 3, 4, 5]:
    print(i)
    
print()
    
dict = {"name": "Aftaab", "age": 25, "city": "New York"}
for key in dict:
    print(key)
    
print()

for value in dict.values():
    print(value)
    
print()

for key, value in dict.items():
    print(key, ":", value)
    
    
print()
print()
print()

print("Nested loops")

for i in range(1, 4):
    for j in range(1, 4):
        print(f"i: {i}, j: {j}")
        
    
```
