```python
### Short circuiting
# Short circuiting is a programming technique where the evaluation of a logical expression is stopped as soon as the result is determined. In Python, the logical operators `and` and `or` exhibit short circuiting behavior.

# For the `and` operator, if the first operand is `False`, the entire expression will be `False`, and the second operand will not be evaluated. For example:
a = False
b = True
result = a and b  # The second operand (b) is not evaluated because a is
# already False
print(result)  # Output: False
# For the `or` operator, if the first operand is `True`, the entire expression will be `True`, and the second operand will not be evaluated. For example:
x = True
y = False
result = x or y  # The second operand (y) is not evaluated because x is
# already True
print(result)  # Output: True

```
