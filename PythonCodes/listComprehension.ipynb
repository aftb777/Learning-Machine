```python
### List Comprehension ###

# List comprehension is a concise way to create lists in Python. It allows you to generate a new list by applying an expression to each item in an iterable, optionally filtering items using a condition.

my_list = []

for char in 'Hello':
    my_list.append(char)
    
print(my_list)
print()

# Using list comprehension, we can achieve the same result in a more concise way:

your_list = [char for char in 'world']
print(your_list)

print()
print()

nums1to10 = [num for num in range(1, 11)]
print(nums1to10)

print()

tableOf2 = [num * 2 for num in range(1, 11)]
print(tableOf2)
```
