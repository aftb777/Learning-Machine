```python
my_set = {1, 2, 3, 4, 5}
print(my_set)

my_set.add(6)
print(my_set)

### Set methods ###
my_set.remove(3)
print(my_set)
my_set.discard(4)
print(my_set)
my_set.clear()
print(my_set)
my_set = {1, 2, 3, 4, 5}
print(my_set)
print(len(my_set))
print(3 in my_set)
print(6 in my_set)
my_set2 = {4, 5, 6, 7, 8}
print(my_set.union(my_set2))
print(my_set.intersection(my_set2))
print(my_set.difference(my_set2))
print(my_set.issubset(my_set2))
print(my_set.issuperset(my_set2))
print(my_set.isdisjoint(my_set2))
```
