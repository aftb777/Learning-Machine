```python
### Ternary Operator
# The ternary operator is a concise way to write an if-else statement in a single line. It is also known as the conditional expression. The syntax for the ternary operator is:
# value_if_true if condition else value_if_false
# Here is an example of using the ternary operator to determine if a number is even or odd:
number = 10
result = "Even" if number % 2 == 0 else "Odd"
print(f"The number {number} is {result}.")
```
