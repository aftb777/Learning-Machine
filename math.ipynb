```python
# Math functions

from cmath import sqrt
from math import factorial


print(round(3.14159, 2))  # Rounds to 2 decimal places

print(round(3.1)) # Rounds to the nearest integer

print(abs(-5))  # Absolute value

print(sqrt(16))  # Square 

print(pow(2, 3))  # 2 raised to the power of 3

print(factorial(5))  # Factorial of 5

print(max(1, 4, 3))  # Maximum of the given numbers

print()
print()

print("### operator precedence ###")

print()

print(2 + 3 * 4)  # Multiplication has higher precedence than addition
```
