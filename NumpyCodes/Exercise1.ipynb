```python
# Nut Butter store sales

import numpy as np
import pandas as pd

np.random.seed(0)

# no of jars sold for 3 different nut butter types in 5 different stores
sales_amount = np.random.randint(20, size=(5, 3))

weekly_sales = pd.DataFrame(sales_amount, index=["Mon", "Tue", "Wed", "Thu", "Fri"], columns=["Almond Butter", "Peanut Butter", "Cashew Butter"])

print(weekly_sales)
```

```python
prices = np.array([10, 8, 12])
butter_prices = pd.DataFrame(prices.reshape(1, 3), index=["Price"], columns=["Almond Butter", "Peanut Butter", "Cashew Butter"])
print(butter_prices)
```

```python
total_sales = weekly_sales.dot(prices)
print(total_sales)
```

```python
print(weekly_sales.shape)
print(prices.shape)
```

```python
weekly_sales["Total Sales"] = total_sales
print(weekly_sales)

print()

print(butter_prices)
```
