```python

# Create a random array and ones array both size (3,5)
random_array = np.random.randint(0, 10, size=(3, 5))
ones_array_35 = np.ones((3, 5))

# Add the two arrays together
print("\nAddition:")
print(random_array + ones_array_35)

# Create another array of ones of shape (5,3)
ones_53 = np.ones((5, 3))

# Try addition
# random_array + ones_53
# Error because shapes (3,5) and (5,3) are not compatible for broadcasting.

# Fix:
ones_35 = np.ones((3, 5))

# Subtract
print("\nSubtraction:")
print(random_array - ones_35)

# Multiply
print("\nMultiplication:")
print(random_array * ones_35)

# Power of 2
print("\nPower using **")
print(random_array ** 2)

# Power using np.square()
print("\nPower using np.square()")
print(np.square(random_array))
```

```python
# Mean
print("\nMean:", np.mean(random_array))

# Maximum
print("Max:", np.max(random_array))

# Minimum
print("Min:", np.min(random_array))

# Standard deviation
print("Standard Deviation:", np.std(random_array))

# Variance
print("Variance:", np.var(random_array))

# Reshape to (3,5,1)
reshaped = random_array.reshape(3, 5, 1)
print("\nReshaped:")
print(reshaped)

# Transpose
transposed = random_array.T
print("\nTransposed:")
print(transposed)
```

```python
# Transpose swaps rows and columns.

# Create arrays for dot product
a = np.random.randint(0, 10, size=(3, 3))
b = np.random.randint(0, 10, size=(3, 2))

print("\nDot Product:")
print(np.dot(a, b))

# Create two arrays of size (4,3)
c = np.random.randint(0, 10, size=(4, 3))
d = np.random.randint(0, 10, size=(4, 3))

# np.dot(c, d) -> Error
# Because inner dimensions do not match.

# Fix using transpose
print("\nDot Product After Transpose:")
print(np.dot(c, d.T))

# Transpose changes d from (4,3) to (3,4),
# allowing (4,3) dot (3,4).
```

```python
# Create two arrays of same shape
arr_a = np.random.randint(0, 10, size=(3, 3))
arr_b = np.random.randint(0, 10, size=(3, 3))

# Compare with >
print("\nGreater Than:")
print(arr_a > arr_b)

# Compare with >=
print("\nGreater Than or Equal:")
print(arr_a >= arr_b)

# Elements greater than 7
print("\nElements > 7:")
print(arr_a[arr_a > 7])

# Equal elements
print("\nEqual Elements:")
print(arr_a == arr_b)
```

```python
# Sort array
print("\nSorted Array:")
print(np.sort(arr_a))

# Sort indexes
print("\nSorted Indexes:")
print(np.argsort(arr_a))
```

```python
# Index of maximum value
print("\nIndex of Maximum:")
print(np.argmax(arr_a))

# Index of minimum value
print("\nIndex of Minimum:")
print(np.argmin(arr_a))

# Indexes of max values along axis=1
print("\nMax Index Along Axis=1:")
print(np.argmax(arr_a, axis=1))

# Indexes of min values along axis=0
print("\nMin Index Along Axis=0:")
print(np.argmin(arr_a, axis=0))
```
